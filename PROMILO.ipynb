```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(GAD['Session Google Ads campaign'], GAD['Engaged sessions'], color='skyblue')
plt.xlabel('Google Ads Campaign')
plt.ylabel('Engaged Sessions')
plt.title('Engaged Sessions by Google Ads Campaign')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.bar(UAB['Age'], UAB['Average engagement time'], color='lightgreen')
plt.xlabel('Age')
plt.ylabel('Average Engagement Time (seconds)')
plt.title('Average Engagement Time by Age Group')
plt.ylim(0, max(UAB['Average engagement time']) * 1.1)  # Adjust y-axis limit for better visualization
plt.tight_layout()
plt.show()
```

```python
UBL
```

```python
# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))
color = 'tab:blue'
bar_width = 0.35  # Define the width of each bar

# Adjust the position of each bar
index = np.arange(len(UBL['Language']))

ax1.bar(index, UBL['Engaged sessions per user'], color=color, width=bar_width, label='Engaged sessions per user')
ax1.set_xlabel('Language')
ax1.set_ylabel('Engaged sessions per user', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticks(index)  # Set x-ticks to align with the bars
ax1.set_xticklabels(UBL['Language'], rotation=45, ha='right')  # Set x-tick labels

ax2 = ax1.twinx()  
color = 'tab:orange'
ax2.plot(index, UBL['Average engagement time'], marker='o', color=color, label='Average engagement time (seconds)')
ax2.set_ylabel('Average engagement time (seconds)', color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Engaged Sessions per User and Average Engagement Time by Language')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sort the data by views in descending order
PSR_sorted = PSR.sort_values(by='Views', ascending=False)

# Take the top 5 entries
top_5 = PSR_sorted.head(5)

# Extracting data for plotting
top_page_paths = top_5['Page path and screen class']
top_views = top_5['Views']
top_avg_time = top_5['Average engagement time']

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting views
ax1.bar(top_page_paths, top_views, color='b', alpha=0.7, label='Views')
ax1.set_ylabel('Views', color='b')
ax1.tick_params('y', colors='b')

# Plotting average time
ax2 = ax1.twinx()
ax2.plot(top_page_paths, top_avg_time, color='r', marker='o', label='Average Time')
ax2.set_ylabel('Average Time', color='r')
ax2.tick_params('y', colors='r')

# Title and labels
plt.title('Top 5 Page paths and screen classes by Views vs Average Time')
plt.xlabel('Page path and screen class')
plt.xticks(rotation=45, ha='right')  # Rotate and align the labels

# Adjust layout to create space between labels
plt.subplots_adjust(bottom=0.3)

plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from the CVR variable
event_names = CVR['Event name']
total_users = CVR['Total users']

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(event_names, total_users, color='b', alpha=0.7)
plt.title('Event Name vs Total Users')
plt.xlabel('Event Name')
plt.ylabel('Total Users')
plt.xticks(rotation=45, ha='right')  # Rotate and align the labels
plt.grid(True)

# Annotate points with their values
for i, txt in enumerate(total_users):
    plt.annotate(txt, (event_names[i], total_users[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()
```
